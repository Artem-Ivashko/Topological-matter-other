```python
import diagonalizer_DSM
from diagonalizer_DSM import diagonalize, pSweep, spectrum_plot
import numpy as np
from copy import copy
from numpy import pi, cos, sin

SitesCount_Default = 400
fsysDefault = diagonalizer_DSM.FinalizedSystem(SitesCount_Default)

#The parameters are taken from the Qi-Zhang RevModPhys for Bi2Se3
ParametersDefault = diagonalizer_DSM.SimpleNamespace( M0= -0.06, M1 = 1.**2, M3 = 0.02, Delta = 3.,\
                     FermiEnergy = 0.0, lBinv2=0., x_shift= (SitesCount_Default - 1.)/2., \
                     EigenvectorsCount = 50, EnergyPrecision = 10**(-5), WavefunctionComponents = 4, \
                     py = 0., pz = 0. )
```

# Somehow, no Fermi arcs! (Look into Pikulin et al papers on Cd3As2 t-b models)

```python
params = copy(ParametersDefault)
pzMin = -0.1
pzMax = 0.1
pzCount = 20
evals, evecs = pSweep(fsysDefault, params, pzMin, pzMax, pzCount, 'pzSweep')
spectrum_plot(evals, pzMin, pzMax, pzCount)

params.pz = 0.03
pyMin = -0.02
pyMax = 0.02
pyCount = 20
evals, evecs = pSweep(fsysDefault, params, pyMin, pyMax, pyCount, 'pySweep')
spectrum_plot(evals, pyMin, pyMax, pyCount)
```
